MW of Support 2 [g mol-1] is showed as object, however it should be numeric

```python
column_MWS2 = Data['MW of Support 2 [g mol-1]']
numeric_column_MWS2 = pd.to_numeric(column_MWS2, errors='coerce')
non_numeric_values = column_MWS2[numeric_column_MWS2.isna()]
display(non_numeric_values)
column_MWS2.loc[non_numeric_values.index] = 0
column_MWS2 = pd.to_numeric(column_MWS2)
```

```python
Data[" MW of Support 2 [g mol-1]"] = column_MWS2
```

All available features

```python
features = Data.loc[:,['Family ', ' Metal Loading [wt.%]',
       'CR Metal [pm]', ' Support 1', 'MW Support 1 [g mol-1]',
       'Name of Support2', 'MW of Support 2 [g mol-1]', 'Name of Support 3',
       'MW Support 3 [g mol-1]', 'Total MW of Support [g mol-1]',
       ' Promoter 1', 'Promoter 1 loading [wt.%]', 'Promoter 2',
       'Promoter 2 loading [wt.%]', 'Calcination Temperature [K]',
       'Calcination duration [h]', 'SBET [m2 g-1]', 'H2/CO2 [-]',
       'GHSV [cm3 h-1 gcat-1]', 'Catalyst amount [g]', 'Pressure [Mpa]',
       'Temperature [K]']]
```

features to be used for the bubble plot

```python
X_unnamed = features.loc[:, [' Metal Loading [wt.%]', 'Promoter 1 loading [wt.%]',
       'Promoter 2 loading [wt.%]', 'Calcination Temperature [K]',
       'Calcination duration [h]', 'SBET [m2 g-1]', 'H2/CO2 [-]',
       'GHSV [cm3 h-1 gcat-1]', 'Catalyst amount [g]', 'Pressure [Mpa]',
       'Temperature [K]']]
```

Calculate the percentages of support, as mentioned in the paper, "the active phase, promoter, and support was normalized to 100% by weight".

```python
X_unnamed['Support [wt.%]'] = 100 - X_unnamed[' Metal Loading [wt.%]'] - X_unnamed['Promoter 1 loading [wt.%]'] - X_unnamed['Promoter 2 loading [wt.%]']
```

```python
X_unnamed.head()
```

Rename the variables for the bubble-plot

```python
X = X_unnamed.rename(columns={' Metal Loading [wt.%]': 'M', 'Promoter 1 loading [wt.%]': 'Pr1', 'Promoter 2 loading [wt.%]': 'Pr2',
                              'Calcination Temperature [K]': 'CT', 'Calcination duration [h]': 'CD', 'SBET [m2 g-1]': 'SBET', 'H2/CO2 [-]': 'H2/CO2',
                              'GHSV [cm3 h-1 gcat-1]': 'GHSV', 'Catalyst amount [g]': 'MC', 'Pressure [Mpa]': 'P',
                              'Temperature [K]': 'T', 'Support [wt.%]': 'S'})
```

Checking if there is null values

```python
print(X.isnull().sum())
```

impute with mode values as mentioned in the paper.

```python
X['MC'] = X['MC'].fillna(X['MC'].mode()[0])
```

```python
X.tail()
```

change the order of the colums for the desired bubble plot

```python
order_col = ['M','S', 'Pr1', 'Pr2', 'SBET','H2/CO2', 'GHSV', 'MC','P', 'T', 'CT', 'CD']
X = X[order_col]
X.head()
```

evaluating the correlations between features

```python
corr = X.corr(method="pearson")
corr.head()
```

Colors and sizes are two variables in the correlation matrix scatter plot

```python
colors = corr.values.flatten()
sizes = np.abs(colors) * 100
```

Now Y axis is in reverse order, so we procced by reversing Y order

```python
x_pos = np.arange(len(corr.columns))
y_pos = np.arange(len(corr.columns))[::-1]
```

```python
X_a, Y_a = np.meshgrid(x_pos, y_pos)
```

```python
X_a = X_a.flatten()
Y_a = Y_a.flatten()
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter( X_a, Y_a, s=sizes*7, c=colors, cmap='plasma_r',alpha=0.8,edgecolors='grey', linewidth=0.4)

ax = plt.gca()
ax.set_xticks(x_pos)
ax.set_yticks(y_pos)
ax.set_xticklabels(corr.columns, rotation=45)
ax.set_yticklabels(corr.columns,rotation=45)
n = len(corr.columns)
ax.set_xticks(np.arange(0.5, n-0.5, 1), minor=True)
ax.set_yticks(np.arange(0.5, n-0.5, 1), minor=True)
ax.grid(which='minor',color='grey', linestyle='-', linewidth=1)
ax.grid(which='major', linewidth=0)

plt.title('Pearsonâ€™s correlation matrix for the set of descriptors (input features)')
plt.colorbar(scatter, label='Correlation Coefficient')
plt.tight_layout()
plt.show()
```
